```python
import pandas as pd
df=pd.read_csv('final_with_mask_mandate.csv')
#df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Education vs Deaths

```python
plt.figure(figsize=(13,8))
sns.scatterplot(x="Edu. < Highschool %", y="COVID-19 Deaths",
            size='Edu. < Highschool %',
            sizes=(50, 200), palette="Dark2", data=df)

plt.xlabel("Education < Highschool %", size=15)
plt.ylabel("Deaths", size=15)
plt.xticks(rotation=75,size=13)
plt.title("Education < Highschool vs Deaths",size=10, weight='bold')
```

## Medical Conditions vs Deaths

```python
plt.figure(figsize=(13,8))
sns.scatterplot(x="Condition Group", y="COVID-19 Deaths",
            hue="Condition", size="Condition",
            sizes=(50, 200), palette="Dark2", data=df)

plt.xlabel("Condition Group", size=15)
plt.ylabel("Fatalities", size=15)
plt.xticks(rotation=75,size=13)
plt.title("Medical Condition vs Deaths",size=10, weight='bold')
```

# Age group vs total deaths

```python
plt.figure(figsize=(13,8))
sns.scatterplot(x="State", y="COVID-19 Deaths",
           hue="Age Group",size='Age Group',
            sizes=(50, 200), palette="Dark2", data=df)

plt.xlabel("State", size=15)
plt.ylabel("Deaths", size=15)
plt.xticks(rotation=75,size=13)
plt.title("State vs Deaths vs Age groups",size=10, weight='bold')
```

## Poverty vs Covid Deaths

```python
plt.figure(figsize=(13,8))
sns.scatterplot(x="poverty %", y="COVID-19 Deaths",
           hue="Age Group",size='Age Group',
            sizes=(50, 200), palette="Dark2", data=df)

plt.xlabel("Poverty %", size=15)
plt.ylabel("Deaths", size=15)
plt.xticks(rotation=75,size=13)
plt.title("Poverty vs Deaths vs Age groups",size=10, weight='bold')
```

## Examining target variable

```python
from scipy.stats import norm

plt.figure(figsize=(10,8))
sns.distplot(df['COVID-19 Deaths'], fit=norm)
plt.title("Target variable distribution plot",size=15, weight='bold')
```
